```python
cat_col=df.select_dtypes(include='object')
for i in cat_col:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df_num = df.select_dtypes(include=['float','int'])

for col in df_num.columns:
    df.fillna(df.mean(numeric_only=True), inplace=True)

```

```python
df.isnull().sum()
#missing values have been filled 
```

```python
# b. Outlier Treatment

def outlier(data,columns):
    for i in columns:
        Q1 = data[i].quantile(0.25)
        Q3 = data[i].quantile(0.75)
        IQR = Q3 - Q1
        threshold = 1.5
        lower_bound= Q1-threshold* IQR
        upper_bound= Q3+threshold* IQR
        data[i]=data[i].clip(lower=lower_bound,upper=upper_bound)
```

```python
outlier(df,df_num)
```

```python
#hyperparameter tuning
from sklearn.cluster import KMeans

inertias = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    result = kmeans.fit(df_scaled)
    inertias.append(result.inertia_)
inertias
```

```python
plt.figure(figsize=(12, 8))
plt.plot(range(1, 11), inertias, "bo-")
plt.xlabel("Value of K")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_scores = []

for k in range(2,11):
    kmeans = KMeans(n_clusters=k, init='random', random_state=42)
    result = kmeans.fit(df_scaled)
    score = silhouette_score(df_scaled, result.labels_)
    silhouette_scores.append(score)

silhouette_scores
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, "bo-")
plt.xlabel("Value of K")
plt.ylabel("Silhouette Score")
plt.show()
```

4. Expected Insights: 
# a. Identification of distinct patient groups based on health and lifestyle attributes.

By analyzing the clustering results, we can identify three main patient groups based on health and lifestyle features:

Low-Risk Group:
Younger patients with healthy BMI, normal cholesterol, and no diabetes or hypertension.
Non-smokers with active lifestyles,this group has the lowest likelihood of chronic diseases.

    
Moderate-Risk Group:Middle-aged patients with slightly elevated cholesterol and blood pressure,Some may have a history of smoking, prediabetes, or mild obesity (BMI 25-30).
Lifestyle changes could help prevent further health issues.


    
High-Risk Group:Older patients (50+ years) with obesity, high cholesterol, diabetes, and hypertension,Likely to have heart disease, high glucose levels, and a history of smoking.
This group needs immediate medical intervention and regular monitoring.


# b. Understanding whether certain clusters correspond to high-risk vs. low-risk patients.
 Cluster 1 (Low-Risk Patients):Mostly younger, active individuals with healthy vitals.
Preventive healthcare is recommended.


    
Cluster 2 (Moderate-Risk Patients):Patients with some risk factors like prediabetes, mild hypertension, or smoking.
Lifestyle interventions (diet, exercise, quitting smoking) could help.


Cluster 3 (High-Risk Patients):Elderly individuals with chronic conditions like hypertension, diabetes, and high cholesterol.
Requires urgent medical care, medication, and monitoring.




# c. Comparison of clustering algorithms to determine which provides the most meaningful segmentation.   


GMM provides the best clustering performance, capturing more complex relationships between patient attributes.It’s useful when patients don’t fit perfectly into distinct categories but share mixed risk factors.

K-Means is a close second, offering fast and interpretable results.

Hierarchical Clustering gives a useful tree structure but is slightly less effective for this dataset.
