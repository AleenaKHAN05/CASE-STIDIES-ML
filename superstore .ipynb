```python
#c. How do monthly or yearly sales trends look over time?
#bivariate analysis bar chart
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Year']=df['Order Date'].dt.year
df['Month']=df['Order Date'].dt.month
msales=df.groupby(['Year','Month'])['Sales'].sum().reset_index()
msales

```

```python
#d. What are the peak and low sales months for each year?
#hue
#double grouping phle year phir month
#lineplot
#d. What are the peak and low sales months for each year?
pksale=msales.loc[msales.groupby('Year')['Sales'].idxmax()]
pksale
```

```python
 lsale=msales.loc[msales.groupby('Year')['Sales'].idxmin()]
 lsale
```

```python
 #LINE PLOT BETWEEN YEAR AND SALES OF PEAKSALE AND LOWSALE
 plt.plot(pksale['Year'],pksale['Sales'],color='#006A67',marker='o',label='Peak sale')
 plt.plot(lsale['Year'],lsale['Sales'],color='#AE445A',marker='*',label='Low sale')
 plt.xlabel('Year')
 plt.ylabel('Sales')
 plt.title("Peak sale's vs. Low sale's")
 plt.legend()
 plt.grid()
 plt.show()
```
